```python
import torch
from fairseq.models.bart import BARTModel
```

```python
bart = torch.hub.load('pytorch/fairseq', 'bart.base')
```

```python
bart_mt_checkpoint_path = (
    "/fs1/groups1/gac50547/edison/data/"
    "bart-mt/semeval_synthetic/checkpoints/checkpoint_best.pt"
)
```

```python
bart_mt_data_bin_path = (
    "/fs1/groups1/gac50547/edison/data/"
    "bart-mt/semeval_synthetic/data-bin/"
)
```

```python
bart = BARTModel.from_pretrained(
        "/".join(bart_mt_checkpoint_path.split("/")[:-1]),
        checkpoint_file=bart_mt_checkpoint_path.split("/")[-1],
        data_name_or_path=bart_mt_data_bin_path,
)
```

```python
bart.eval()
```

```python
# example = "Iceland PM Calls Snap Vote as Pedophile Furor Crashes <mask>"
# example = "In an apparent first <mask> Iran and Israel <mask> each other militarily"
#example = "France is ‘ hunting down its citizens who <mask> ’ without trial in Iraq"
example = "The Latest : Japan praises U.S. Korean missile <mask>"
```

```python
bart.fill_mask(['The cat <mask> on the <mask>.'], topk=3, beam=10)
```

```python
processed_example = example.replace("<mask>", " <mask> ").replace("  ", " ")
```

```python
bart.encode(processed_example)
```

```python
hypotheses_batch = bart.sample(
    [example],
    beam=5,
    topk=5,
    match_source_len=False,
    min_len=10,
)
```
